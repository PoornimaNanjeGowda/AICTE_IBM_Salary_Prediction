```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, OneHotEncoder

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

models = {
    "LogisticRegression": LogisticRegression(),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier()
}

results = {}

for name, model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

# Create a list of colors (must match number of algorithms)
colors = ['skyblue', 'lightgreen', 'salmon', 'plum', 'orange', 'cyan']

# Ensure color list matches number of bars
color_list = colors[:len(results)]

# Plotting
plt.figure(figsize=(5, 5))
plt.bar(results.keys(), results.values(), color=color_list)
plt.ylabel('Accuracy Score')
plt.title('Model Comparison')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Define models
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier()
}

results = {}

# Train and evaluate
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    results[name] = acc
    print(f"{name}: {acc:.4f}")

# Get best model
best_model_name = max(results, key=results.get)
best_model = models[best_model_name]
print(f"\n Best model: {best_model_name} with accuracy {results[best_model_name]:.4f}")

# Save the best model
joblib.dump(best_model, "best_model.pkl")
print(" Saved best model as best_model.pkl")
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import joblib

# Load dataset
df = pd.read_csv("/content/adult 3 IBM INTERNSHIP.csv")

# Select and rename relevant columns
df = df[["age", "education", "occupation", "gender", "hours-per-week", "income"]].dropna()
df.rename(columns={
    "occupation": "job_title",
    "gender": "sex",
    "hours-per-week": "hours_per_week",
    "income": "salary"
}, inplace=True)

# Encode categorical features
label_encoders = {}
for col in ["education", "job_title", "sex", "salary"]:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Features and target
X = df[["age", "education", "job_title", "sex", "hours_per_week"]]
y = df["salary"]

# Save the training column names
trained_columns = X.columns.tolist()
joblib.dump(trained_columns, "trained_columns.pkl")

# Save the label encoders
joblib.dump(label_encoders, "label_encoders.pkl")

# Train the model
model = RandomForestClassifier()
model.fit(X, y)

# Save the trained model
joblib.dump(model, "best_model.pkl")
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import joblib

# Load your cleaned dataset
df = pd.read_csv("/content/adult 3 IBM INTERNSHIP.csv")

# Rename relevant columns to standard form
df.rename(columns={
    "occupation": "job_title",
    "hours-per-week": "hours_per_week",
    "sex": "gender",
    "income": "salary"
}, inplace=True)

#  Add a random or default 'experience' column if not in dataset
# You can replace this with real data if available
if "experience" not in df.columns:
    df["experience"] = df["age"] - 22
    df["experience"] = df["experience"].clip(lower=0, upper=40)

# Convert classification labels to numeric regression target
df["salary"] = df["salary"].map({">50K": 60000, "<=50K": 40000})

# Encode categorical variables
label_encoders = {}
for col in ["education", "job_title", "gender"]:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

#  Define features and target
X = df[["age", "education", "job_title", "gender", "experience", "hours_per_week"]]
y = df["salary"]

# Train model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)

# Save model and metadata
joblib.dump(model, "/content/best_model.pkl.gz")
joblib.dump(X.columns.tolist(), "trained_columns.pkl")
joblib.dump(label_encoders, "label_encoders.pkl")
```

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
import joblib

#  Step 1: Load dataset
df = pd.read_csv("/content/adult 3 IBM INTERNSHIP.csv")

#  Step 2: Clean column names (optional)
df.columns = df.columns.str.strip().str.lower().str.replace("-", "_")

#  Step 3: Define features and targets
features = ["age", "education", "gender","occupation", "hours_per_week"]

# Create targets
df["salary_class"] = df["income"].str.strip()  # classification label
df["salary_amount"] = df["hours_per_week"] * 1000  # mock regression label

X = df[features]
y_class = df["salary_class"]
y_reg = df["salary_amount"]

#  Step 4: Encode categorical features
X_encoded = pd.get_dummies(X)

# Save column names for model alignment
feature_cols = X_encoded.columns

#  Step 5: Split data
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_encoded, y_class, test_size=0.2, stratify=y_class, random_state=42
)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_encoded, y_reg, test_size=0.2, random_state=42
)

#  Step 6: Train models
clf_model = RandomForestClassifier(n_estimators=100, random_state=42)
clf_model.fit(X_train_clf, y_train_clf)

reg_model = RandomForestRegressor(n_estimators=100, random_state=42)
reg_model.fit(X_train_reg, y_train_reg)

#  Step 7: Save models
clf_model.feature_names_in_ = feature_cols
reg_model.feature_names_in_ = feature_cols

joblib.dump(clf_model, "/content/clf_model.pkl")
joblib.dump(reg_model, "/content/reg_model.pkl")

print(" Models saved: clf_model.pkl and reg_model.pkl")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
import joblib
import plotly.express as px
import plotly.graph_objects as go
import io

# Load model and encoders
model = joblib.load("/content/best_model.pkl.gz")
trained_columns = joblib.load("/content/trained_columns.pkl")
label_encoders = joblib.load("/content/label_encoders.pkl")

# USD to INR conversion rate
USD_TO_INR = 83.5

# Page config
st.set_page_config(page_title=" Salary Predictor", layout="centered")
st.title(" Employee Salary Predictor")
st.markdown("Estimate your salary based on your profile.")

# --- Custom CSS styling ---
custom_css = """
<style>
[data-testid="stAppViewContainer"] {
    background: linear-gradient(to right, #0f2027, #203a43, #2c5364);
    padding: 2rem;
    color: white;
}
.block-container {
    backdrop-filter: blur(10px);
    background-color: rgba(0, 0, 0, 0.3);
    border-radius: 15px;
    padding: 2rem;
}
h1, h2, h3 {
    color: #ffffff;
    text-shadow: 1px 1px 2px #000;
}
.stButton > button, .stDownloadButton > button {
    background-color: #2c3e50;
    color: white;
    font-weight: bold;
    border-radius: 8px;
    padding: 10px 20px;
    border: none;
    box-shadow: 2px 2px 4px #00000055;
}
</style>
"""
st.markdown(custom_css, unsafe_allow_html=True)

# --- Tabs ---
tabs = st.tabs([" Home", " Predict Salary", " Upload & Predict"])

# --- Home Tab ---
with tabs[0]:
    st.title(" Welcome to Salary Prediction App")
    st.markdown("""
        This app helps you predict:
        -  Annual and Monthly Salary
        -  For Individuals or Bulk CSV Upload

        Powered by a trained **Random Forest** model (Accuracy: 0.84) using features like:
        -  Age
        -  Education Level
        -  Job Title
        -  Gender
        -  Hours Worked Per Week
    """)

# --- Individual Prediction Tab ---
with tabs[1]:
    st.title(" Predict Your Salary")

    age = st.number_input("Enter your age", min_value=18, max_value=70, step=1)
    education = st.selectbox("Select your education level", label_encoders["education"].classes_)
    job_title = st.selectbox("Select your job title", label_encoders["job_title"].classes_)
    gender = st.radio("Select your gender", label_encoders["gender"].classes_)
    experience = st.slider("Years of experience", min_value=0, max_value=50, value=2)
    hours = st.slider("Hours worked per week", min_value=1, max_value=100, value=40)

    # Predict button
    if st.button(" Predict Salary"):
        try:
            # Encode inputs
            input_data = pd.DataFrame([{
                "age": age,
                "education": label_encoders["education"].transform([education])[0],
                "job_title": label_encoders["job_title"].transform([job_title])[0],
                "gender": label_encoders["gender"].transform([gender])[0],
                "experience": experience,
                "hours_per_week": hours
            }])
            input_data = input_data.reindex(columns=trained_columns)

            # Prediction
            annual_usd = model.predict(input_data)[0]
            annual_usd = np.clip(annual_usd, 18000, 90000)  # Clamp extreme values
            annual_inr = annual_usd * USD_TO_INR
            monthly_usd = annual_usd / 12
            monthly_inr = monthly_usd * USD_TO_INR

            # Display results
            st.warning("  Disclaimer: This salary prediction is an estimate based on provided data and does not guarantee actual compensation.Results may vary depending on market trends, company policies, and individual qualifications.")
            st.success(f" **Annual Salary:** {annual_inr:,.2f} ( ${annual_usd:,.2f})")
            st.info(f" **Monthly Salary:** {monthly_inr:,.2f} ( ${monthly_usd:,.2f})")

            # Salary Breakdown Chart
            experience_range = list(range(0, 21))
            salary_growth = [monthly_inr * (1 + 0.05 * i) for i in experience_range]

            fig = px.line(
                x=experience_range,
                y=salary_growth,
                markers=True,
                labels={"x": "Years of Experience", "y": "Predicted Monthly Salary (INR)"},
                title=" Salary Growth Over Time"
            )
            st.plotly_chart(fig)




            # Generate Report
            report = io.StringIO()
            report.write(" Salary Prediction Report\n")
            report.write("-----------------------------\n")
            report.write(f" Age: {age}\n")
            report.write(f" Education: {education}\n")
            report.write(f" Job Title: {job_title}\n")
            report.write(f" Gender: {gender}\n")
            report.write(f" Hours per week: {hours}\n\n")
            report.write(" Salary Predictions:\n")
            report.write(f"- Estimated Annual Salary: {annual_inr:,.2f}\n")
            report.write(f"- Estimated Monthly Salary: {monthly_inr:,.2f}\n")

            st.download_button(" Download Salary Report", report.getvalue(), file_name="salary_report.txt")

        except Exception as e:
            st.error(f" Prediction failed: {e}")

# --- Upload Prediction Tab ---
with tabs[2]:
    st.title(" Upload CSV to Predict Multiple Salaries")

    st.markdown("""
    Upload a CSV file with the following columns:
    - `age`, `education`, `job_title`, `gender`, `experience`, `hours_per_week`
    """)

    uploaded_file = st.file_uploader("Choose a CSV file", type=["csv"])

    if uploaded_file is not None:
        try:
            df = pd.read_csv(uploaded_file)

            for col in ["education", "job_title", "gender"]:
                if col in df.columns:
                    df[col] = label_encoders[col].transform(df[col])

            df = df.reindex(columns=trained_columns)

            predictions = model.predict(df)
            predictions = np.clip(predictions, 18000, 90000)
            df["Predicted_Annual_Salary_INR"] = predictions * USD_TO_INR
            df["Predicted_Monthly_Salary_INR"] = df["Predicted_Annual_Salary_INR"] / 12

            st.success(" Predictions generated successfully.")
            st.dataframe(df)

            download = df.to_csv(index=False).encode("utf-8")
            st.download_button(" Download Predictions CSV", data=download, file_name="salary_predictions.csv", mime="text/csv")

        except Exception as e:
            st.error(f"Something went wrong: {e}")

# --- Footer ---
footer = """
<br><hr>
<div style='text-align: center; color: #ccc; font-size: 1.1rem; padding: 10px;'>
     2025 Salary Prediction App | Built with  by Poornima N | Powered by Streamlit and scikit-learn
</div>
"""
st.markdown(footer, unsafe_allow_html=True)
```
